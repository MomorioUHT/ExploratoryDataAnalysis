### 12 สร้าง Column ใหม่ที่ชื่อว่า "Fatalities Percent"<br/>

### 15 คัดเลือก 7 สายการบิน (Operator)
ดังนี้ ออกมา ['All Nippon Airways', 'EVA Air', 'Scaandinavian Airlines (SAS)', 'Singapore Airlines', 'Swisssair','Thai Airways', 'United Air Lines']
ซึ่งชื่อทั้งหมดอยู่ใน star7 csv แล้วตั้งชื่อว่า star_7 แล้วแสดงข้อมูล

```python
star_aliance = pd.read_csv('star_alliance_7.csv')
star_7 = df.loc[df['Operator'].isin(star_aliance['Star Alliance'])]
star_7.head()
```

### 16 สรุปรวมจำนวนอุบัติเหตุที่เกิดขึ้นของแต่ละสายการบินใน star_7

```python
star_7.groupby('Operator').size()
```

### 17 นำอุบัติเหตุของการบินไทย (Thai Airways) มาเรียงลำดับจากล่าสุดย้อนกลับไป

```python
star_7['Date'].isna().sum()
```

```python
star_7['Date'] = pd.to_datetime(star_7['Date'])
star_7['Date'].dtype
```

```python
star_7.loc[star_7['Operator'] == 'Thai Airways'].sort_values('Date', ascending=False)
```

### 18 แสดงข้อมูลของ 7 สายการบินเรียงตาม Operator จากน้อยไปมาก และในแต่ละ Operator เรียงจากอุบัติเหตุที่เกิดขึ้นก่อน มายังครั้งล่าสุด

```python
star_7.sort_values(['Operator', 'Date'], ascending=[True, True])
```

### 19 คัดเลือกอุบัติเหตุของสายการบิน Thai Airways ในทศวรรษที่ 90 (1990-1999) ออกมา

```python
star_7.loc[(star_7['Date'].dt.year >= 1990) & (star_7['Date'].dt.year <= 1999)]
```

### 20 ในตาราง star_7 สกัดคำแรกของคอลัมฯ์ 'AC Type' ออกมาเป็นชื่อบริษัทที่สร้างเครื่องบิน
จากนั้นนับจำนวนเครื่องบินที่ตก แยกตามสายการบินและบริษัทเครื่องบิน, ไม่ต้องสนใจข้อมูลที่ผิดพลาด

```python
star_7['Company'] = star_7['AC Type'].str.split().str.get(0)
star_7.pivot_table(index='Company',
                   columns='Operator',
                   values='Date', #any values, this is for counting only!
                   aggfunc='size')
```

### 21 นับจำนวนคนตายของ 7 สายการบินแยกตามทศวรรษ และสายการบิน

```python
star_7['Decade'] = star_7['Date'].dt.year//10*10 # Cheat
star_7['Decade']
```

```python
star_7.pivot_table(index='Operator',
                   columns='Decade',
                   values='Fatalities',
                   aggfunc='sum')
```

### 22 เปลี่ยน Column Operator ของ star_7 ให้เป็น Category โดยเรียงลำดับตามตัวอักษร

```python
pd.Categorical(star_7['Operator'], 
               categories=['All Nippon Airways',
                            'EVA Air',
                            'Scandinavian Airlines (SAS)',
                            'Singapore Airlines',
                            'Swissair',
                            'Thai Airways',
                             'United Airlines']
                , ordered=True)
```

### 23 แสดงแผนภาพ Histogram และ Rugplot ของ Fatalities Percent ของอุบัติเหตุจาก 7 สายการบิน

```python
sns.histplot(y=star_7['Fatalities Percent'], x=star_7['Operator'])
```

### 24 แสดงกราฟแท่งแสดงจำนวนคนตายรวมของ 7 สายการบิน

```python
deadstats = star_7.groupby('Operator', as_index=False)['Fatalities'].sum()

fig, ax = plt.subplots(figsize=(10, 5))
ax.set_title("Fatalities count by Airlines")

bars = ax.bar(
    deadstats['Operator'],
    deadstats['Fatalities']
)

ax.bar_label(bars, labels=[f'{x:.0f} People(s)' for x in deadstats['Fatalities']], padding=3)

ax.set_xlabel("Airlines")
ax.set_ylabel("Dead Count")
plt.xticks(rotation=45)

plt.show()
```
