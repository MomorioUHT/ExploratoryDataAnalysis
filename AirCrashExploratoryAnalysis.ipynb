### 12 สร้าง Column ใหม่ที่ชื่อว่า "Fatalities Percent"<br/>
แสดงสัดส่วนจำนวนคนตายรวมต่อจำนวนคนในเครื่องบิน (Abord) และแสดง 5 แถวแรกของตารางนี้

```python
df['Fatalities Percent'] = (df['Fatalities'] / df['Aboard']) * 100
df.head()
```

### 13 แสดงข้อมูล 5 แถวแรกของอุบัติเหตุที่มีคนตายมากที่สุดเรียงจากมากไปน้อย

```python
df1 = df.sort_values('Fatalities', ascending=False)
df1.head(5)
```

### 14 แสดงข้อมูล 5 แถวแรกของอุบัติเหตุที่มี Fatalities Percent สูงที่สุดเรียงจากมากไปน้อย<br/>
